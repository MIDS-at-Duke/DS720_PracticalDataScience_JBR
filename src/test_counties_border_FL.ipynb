```python
import pandas as pd
import numpy as np
from pyspark.sql import SparkSession

counties = pd.read_csv(
    "https://github.com/wpinvestigative/arcos-api/raw/master/data/county_fips.csv"
)
```

```python
al_geo = counties[(counties["BUYER_STATE"] == "AL") | (counties["BUYER_STATE"] == "GA")]
```

```python
border_counties = {
    "GA": "SEMINOLE DECATUR GRADY THOMAS BROOKS LOWNDES ECHOLS CLINCH WARE CHARLTON CAMDEN".split(),
    "AL": "BALDWIN ESCAMBIA COVINGTON GENEVA HOUSTON".split(),
}
```

```python
# Create a SparkSession
spark = SparkSession.builder.getOrCreate()

# Define the file path
file_path = "../../arcos_all_washpost.tsv"  # Change to your file path

# Read the data
df = spark.read.csv(file_path, sep="\t", header=True)
```

```python
# Filtering data using border_counties keys to match BUYER_STATE and values to match BUYER_COUNTY

filtered_df = df.filter(
    (df.BUYER_STATE == "AL") & (df.BUYER_COUNTY.isin(border_counties["AL"]))
)
filtered_df.write.parquet("../../AL_border.parquet")

filtered_df = df.filter(
    (df.BUYER_STATE == "GA") & (df.BUYER_COUNTY.isin(border_counties["GA"]))
)
filtered_df.write.parquet("../../GA_border.parquet")
```
